```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier

csv = pd.read_csv('Employee_Int_Convert.csv')

X = csv.drop(['LeaveOrNot'], axis = 1)
y = csv['LeaveOrNot']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42)

# the classifier
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)

features = list(csv.columns)

features.remove('LeaveOrNot')

export_graphviz(tree, out_file="tree.dot", class_names=["leave", "stay"],
                feature_names=features, impurity=False, filled=True)
import graphviz
with open("tree.dot") as f:
    dot_graph = f.read()
display(graphviz.Source(dot_graph)) 
```
